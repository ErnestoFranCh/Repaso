## **`DataFrame.sort_values()`**



```python
#sintax
DataFrame.sort_values(
    by,
    axis=0,
    ascending=True,
    inplace=False,
    kind='quicksort',
    na_position='last',
    ignore_index=False
)
```
- `by`:	Nombre de la columna (o lista de columnas) por la que quieres ordenar.

- `ascending`:	`True` (por defecto): orden ascendente; `False`: descendente. Acepta lista si usas varias columnas.

- `inplace`:	`False` por defecto. Si es `True`, modifica el DataFrame original.

- `na_position`:	`'last'` (por defecto): los `NaN` van al final. Usa `'first'` si los quieres al inicio.

- `ignore_index`:	Si es `True`, reinicia los índices después de ordenar.

- `kind`:	Algoritmo de ordenamiento: `'quicksort'`, `'mergesort'`, `'heapsort'`.

##`DataFrame.sort_index()`

Ordena un DataFrame por su índice, ya sea filas o columnas.



```python
#sintax
DataFrame.sort_index(
    axis=0,
    level=None,
    ascending=True,
    inplace=False,
    kind='quicksort',
    na_position='last',
    sort_remaining=True,
    ignore_index=False,
    key=None
)
```
- `axis`:	`0` (por defecto): ordena filas por índice; `1`: ordena columnas por sus nombres.

- `ascending`	`True` (por defecto): orden ascendente; `False`: descendente.

- `inplace`	`False` por defecto. Si es `True`, modifica el `DataFrame` original.

- `level`	Si el índice es jerárquico (`MultiIndex`), permite ordenar por un nivel específico.

- `ignore_index`	Si es `True`, reinicia los índices después de ordenar.

- `key`	Función para transformar los valores del índice antes de ordenarlos (útil para personalizar el orden).

## `DataFrame.merge()`



```python
#sintaxis
pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,
         left_index=False, right_index=False, suffixes=('_x', '_y'))

```
- `left`, `right`:	Los `DataFrames` a combinar.
    - El `DataFrame` que pongas en `left` es la base en las uniones con `pd.mere()`.

- `how`:	Tipo de unión: `'inner'`, `'outer'`, `'left'`, `'right'`.
    - `inner`: Solo filas con claves coincidentes en ambos `DataFrames`.
    - `'left'` : Todas las filas del DataFrame izquierdo, y lo que coincida del derecho.
    - `'right'`: Todas las filas del derecho, y lo que coincida del izquierdo.
    - `'outer'`: Todas las filas de ambos DataFrames, uniendo donde se pueda.

- `on` :	Columna(s) en común (mismo nombre en ambos).

- `left_on`, `right_on`	Si las columnas clave tienen nombres diferentes.

- `left_index`, `right_index`	Usa los índices como clave de unión.

- `suffixes` :	Sufijos para columnas duplicadas (`_x`, `_y` por defecto)


````python
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Nombre': ['Ana', 'Luis', 'María']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Edad': [23, 35, 40]
})
````

````python
#how = left
pd.merge(df1,df2,how ='left',on='ID')
#Todas las de la izquierda y lo que coincidad del derecho
````

````python
#how = inner

pd.merge(df1,df2,how='inner',on='ID') #Solo filas coincidentes
````

````python
#how = right
pd.merge(df1,df2,how ='right',on='ID')
````

````python
#how = outer
pd.merge(df1,df2,how='outer',on='ID')
````

````python
df3 = pd.DataFrame({
    'ID_name': [1, 2, 3],
    'Nombre': ['Ana', 'Luis', 'María']
})

df4 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Edad': [23, 35, 40]
})
````

````python
pd.merge(df3,df4,how='left',left_on = 'ID_name',right_on='ID')
````

````python
df5 = pd.DataFrame({
    'ID': [101, 102, 103],
    'Nombre': ['Ana', 'Luis', 'María']
})

df6 = pd.DataFrame({
    'Edad': [25, 30],
    'Ciudad': ['CDMX', 'Guadalajara']
}, index=[101, 103])
````

````python
pd.merge(df5,df6,how='left',left_on='ID',right_index=True)
````

## `DataFrame.groupby()`

Agrupa las filas de un DataFrame según los valores de una o más columnas. Luego puedes aplicar funciones como `.sum()`, `.mean()`, `.count()`, etc., a cada grupo.



```python
df.groupby('columna').función() # 1 columna

df.groupby(['col1', 'col2',..]).función() #multiples col


```

````python
EmployeeData.head(2)
````

````python
EmployeeData.groupby('Department')['Annual Salary (USD)'].sum()
#aqui se muestra la suma de salario por departamento
````

| Función              | Qué hace                                   |
|----------------------|---------------------------------------------|
| `.sum()`             | Suma los valores de cada grupo              |
| `.mean()`            | Calcula el promedio                         |
| `.count()`           | Cuenta los registros                        |
| `.min()` / `.max()`  | Mínimo / máximo                             |
| `.agg()`             | Permite aplicar múltiples funciones         |
| `.size()`            | Cuenta el tamaño de cada grupo              |
| `.first()` / `.last()` | Primer / último valor del grupo           |


````python
EmployeeData.groupby('Department').agg({
    'Annual Salary (USD)': ['mean','max','min','sum']
})
````

### `lambda` y funciones en `.agg`

````python
EmployeeData.groupby('Department').agg({
    'Annual Salary (USD)' : ['mean',
                             lambda x : x.max()-x.min()]
})
````

Nombres personalizados de las columnas por tuplas.

````python
EmployeeData.groupby('Department').agg(
    #name_col = (col,function)
    Salario_promedio = ('Annual Salary (USD)','mean'),
    Rango_salario = ('Annual Salary (USD)',
                     lambda x : x.max()-x.min())
)
````

````python
def rango(valores):
    return valores.max() - valores.min()

EmployeeData.groupby('Department').agg(
    #name_col = (col,function)
    Salario_promedio = ('Annual Salary (USD)','mean'),
    Rango_salario = ('Annual Salary (USD)',rango)
)
````

## `DataFrame.fillna()`

Rellena los valores faltantes (`NaN`) en un DataFrame con un valor específico, o usando una técnica de propagación (como copiar valores anteriores o siguientes).

```python
df.fillna(value=None, method=None, axis=None, inplace=False, limit=None)

```

- `value`:	El valor con el que quieres reemplazar los NaN. Puede ser un escalar, diccionario o Series.

- `method`:	Método de propagación:
    - `'ffill'` (forward fill) : Copia el valor anterior no nulo hacia abajo.
    - `'bfill'` (backward fill): Copia el valor siguiente no nulo hacia arriba.

- `axis`:	`0` para filas (por defecto), `1` para columnas.

- `inplace`: 	Si es `True`, modifica el `DataFrame` original.

- `limit`:	Número máximo de valores consecutivos a llenar.

## `DataFrame.apply()`

Sirve para aplicar una función a lo largo de un eje (filas o columnas) de un DataFrame.



```python
#sintaxis
DataFrame.apply(func, axis=0, raw=False, result_type=None, args=(), **kwargs)

```
- `func`: la función que quieres aplicar. Puede ser:
    - Una función de Python (`sum`, `len`, `np.mean`, etc.)
    - Una función lambda
    - Una función personalizada que tú definas.

- `axis`:
    - `0` (o `'index'`): aplica la función a lo largo de las columnas (es decir, la función recibe cada columna como una Series)
    - `1` (o `'columns'`): aplica la función a lo largo de las filas

- `raw:`
    - Si es `True`, pasa los valores como un arreglo ndarray sin etiquetas.
    - Si es `False` (default), pasa cada columna o fila como una Series con etiquetas.
- `args=()`: Permite pasar argumentos adicionales a la función que estás aplicando. Debes colocarlos como una tupla.

- `**kwargs`: Te permite pasar argumentos con nombre (keywords) adicionales a tu función.


````python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [10, 20, 30]
})
df
````

````python
df.apply(sum) #por columnas
````

````python
#por filas usando una funcion lambda
df.apply(lambda rows: rows['A']+rows['B'], axis=1)

#similar
# df.apply(sum,axis=1)
````

````python
def rango(series):
  return series.max()-series.min()

df.apply(rango)
````

````python
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df
````

````python
def multiplicar(x, factor):
    return x * factor

df = pd.DataFrame({'A': [1, 2, 3]})

df['A'].apply(multiplicar, args=(10,))
# Esto es equivalente a que Pandas llame: multiplicar(valor_de_A, 10)
````

# Funciones de ventana y de series temporales
- **`DataFrame.rolling()`**: Proporciona operaciones de ventana móvil.
- **`DataFrame.expanding()`**: Proporciona operaciones acumulativas.
- **`DataFrame.resample()`**: Remuestrea series temporales a una frecuencia diferente.
- **`pd.date_range()`**: Genera un rango de fechas.
- **`pd.to_datetime()`**: Convierte una cadena u objeto a tipo datetime.

## `DataFrame.rolling()`

Aplica una función agregada (como la media, la suma, etc.) sobre una ventana de elementos consecutivos en una serie de datos.



```python
#sintaxis
DataFrame.rolling(window, min_periods=None, axis=0, method='single')

```
- `window`: El tamaño de la ventana, que indica el número de observaciones a considerar en cada paso de la ventana deslizante.
    - Ejemplo: `window=4` significa que la operación se realizará sobre 4 filas consecutivas de datos.

- `min_periods` (opcional):  El número mínimo de observaciones que deben estar presentes en la ventana para realizar la operación. Si es menor que el tamaño de la ventana, se llenan los valores faltantes (`NaN`) al inicio de la serie.
    - `min_periods=2` significa que, para aplicar una función como la media, la ventana debe tener al menos 2 valores no nulos.
    
- `method` (opcional):  Define el tipo de ventana a usar. El valor predeterminado es 'single', pero puede ser 'table' si estás aplicando la ventana sobre tablas de datos multidimensionales (poco común).

````python
# Datos de ejemplo: ventas trimestrales de una empresa
data = {'Año': [2021, 2021, 2021, 2021, 2022, 2022, 2022, 2022],
        'Trimestre': ['Q1', 'Q2', 'Q3', 'Q4', 'Q1', 'Q2', 'Q3', 'Q4'],
        'Ventas': [100, 120, 130, 140, 150, 160, 170, 180]}

df = pd.DataFrame(data)
df
````

````python
# Cálculo de la media móvil de 4 trimestres
df['Promedio_Móvil'] = df['Ventas'].rolling(window=4).mean()

print(df)
````

`ventana.funcion` --> `rolling.function`
- La ventana de tamaño 4 (cuatro trimestres) se mueve sobre las columnas `Ventas` para calcular el promedio de cada conjunto de 4 trimestres consecutivos.

- En las primeras tres filas, el promedio móvil no puede calcularse porque no hay suficientes datos anteriores (menos de 4 trimestres).

- A partir del cuarto trimestre, puedes ver el promedio de las ventas de los últimos 4 trimestres.

##`DataFrame.expanding()`

Se usa para aplicar funciones acumulativas sobre los datos desde el inicio hasta la fila actual, es decir, una ventana creciente. A diferencia de `.rolling()`, que usa una ventana fija, `.expanding()` comienza con un número mínimo de observaciones y se expande hacia abajo a medida que avanzamos en el `DataFrame`.



```python
#sintaxis
DataFrame.expanding(min_periods=1, axis=0, method='single')
```

- `min_periods` (por defecto = 1): Número mínimo de observaciones requeridas para que se calcule el resultado. Si no se alcanzan, el resultado será `NaN`

- `axis` (por defecto = 0): Eje sobre el que se aplicará la operación.
    - `0 `→ vertical (por filas),
    - `1` → horizontal (por columnas)

- `method` (por defecto = 'single'):


````python
# Ventas mensuales simuladas
data = {'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
        'Ventas': [100, 120, 150, 130, 160]}

df = pd.DataFrame(data)

df
````

````python
# Promedio acumulado de ventas
df['Promedio_Acumulado'] = df['Ventas'].expanding(min_periods=1).mean()
df
````

- El promedio acumulado se calcula incluyendo todas las ventas desde el inicio hasta el mes actual.

- En la segunda fila: (100 + 120) / 2 = 110.0

- En la tercera fila: (100 + 120 + 150) / 3 = 123.33...

- Así se va expandiendo la ventana hasta el final.

##`DataFrame.resample()`

Permite agrupar datos temporales por frecuencia (diaria, mensual, anual, etc.) y aplicar funciones de agregación como `mean()`, `sum()`, `count()`, etc.

```python
DataFrame.resample(rule,
                   axis=0,
                   closed=None,
                   label=None,
                   convention='start',
                   kind=None,
                   loffset=None, base=None, on=None, level=None)

```
- `rule` (obligatorio): Frecuencia para reagrupar los datos.
    - `'D'`: diario
    - `'W'`: semanal
    - `'M'`: mensual
    - `'Q'`: trimestral
    - `'A'`: anual
    - `'H'`: por hora

- `on` (opcional): Si el índice no es un datetime, puedes usar este parámetro para indicar qué columna contiene las fechas.


````python
# Crear un DataFrame de visitas diarias
fechas = pd.date_range(start="2023-01-01", periods=10, freq='D')
visitas = [120, 135, 150, 160, 180, 145, 155, 170, 190, 210]

df = pd.DataFrame({'Fecha': fechas, 'Visitas': visitas})
df
````

````python
# Reagrupar por semana y sumar visitas
df_semanal = df.resample('W', on='Fecha').sum()

print(df_semanal)
````
