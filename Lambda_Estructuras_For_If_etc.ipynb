# `if`

`if` es una estructura condicional que permite ejecutar un bloque de código solo si se cumple una condición.

```python
#sintaxis

    # código que se ejecuta si la condición es verdadera
```

Python tiene un conjunto de valores que, al evaluarse como booleanos, se consideran `False`:

| Tipo        | Valor considerado False           |
|-------------|-----------------------------------|
| None        | `None`                              |
| Numérico    | 0, 0.0, 0j                         |
| Strings     | `""` (cadena vacía)                 |
| Listas      | `[]` (lista vacía)                  |
| Tuplas      | `()`                                |
| Diccionarios| `{}`                                |
| Sets        | `set()`                             |
| Booleanos   | `False `                            |

````python
x = 0 #False
if not x: # not False = True, se ejectura el if
  print("x es verdadero")
````

## `break` y `continue`

`break`.

¿Qué hace?
- Sale completamente del bucle (aunque la condición aún sea verdadera).

- El programa continúa después del bucle.

`continue`.

¿Qué hace?
- Interrumpe la iteración actual del bucle.

- Pasa directamente a la siguiente vuelta del ciclo, sin ejecutar el resto del bloque.

Ambos se usan en `for` y `while`

````python
#break
x = 0
while x < 10:
    print(x)
    if x == 4:
        break  #  Se detiene aquí
    x += 1
````

````python
x = 0
while x < 5:
    x += 1
    if x == 3:
      print('salta y regresa al inicio del loop')
      continue  #Salta el print solo cuando x == 3
    print(x)
````

## `for`

Se usa para iterar sobre una secuencia de elementos: listas, tuplas, diccionarios, cadenas de texto, conjuntos (set), objetos generados por range(), etc.



``` python
for variable in secuencia:
    # bloque de código que se ejecuta por cada elemento
```



````python
#recorrer listas
pares = [[1,2,3],[3,4],[5,6,3]]

for x,*y in pares:  #el operador * solo se permite dentro de una asignacion multiple: y* in pares: da error
  print(x,y)
````

````python
#Recorrer strings
language = 'Python'
for letter in language:
    print(letter)


for i in range(len(language)):
    print(language[i])
````

````python
#recorrer diccionarios
dic = {'a':1,'b':2,'c':3}

for key in dic.keys():  #for key in dic: da el mismo resultado
  print(key)
print('--------')
for value in dic.values():
  print(value)
print('--------')
for key,value in dic.items():
  print(key,value)
````

## `break` y `continue`


````python
numbers = (1,2,3,4,5)

for number in numbers:
  print(number)
  if number == 3:
    print('aqui para')
    break
````

````python
for number in numbers:
  if number == 3:
    print('aqui continua')
    continue
  print(number)
````

#Compresion de listas.

Es una manera corta de construir listas a partir de iteraciones y condiciones.

Retomemos el `if` en una linea:

```python
# sintaxis
code1 if condition else code1
```
si la condicion se cumple se devuelve `code1`.



```python
lst = [expresión for item in iterable]
```
Similar a la estructura del `if`, `expresion` es el item que se guarda en la lista


````python
palabra = 'Python'
lst = list(palabra)
print(lst)

lst1 = [i for i in palabra]
print(lst1)
````

````python
#algunos otros ejemplos:

l = [i-1 for i in range(10)]
print(l)

l1 = [(i,i**i) for i in range(10)]
print(l1)

l2 = [[i,i+1,i+2] for i in range(5)]
print(l2)

l3 = [{i,i-1,i+1} for i in range(5)]
print(l3)
````

## Listas filtradas.

Recuerda como funciona:

```python
# sintaxis
code1 if condition else code1
```

si la condicion se cumple se devuelve `code1`.

###Lista filtrada
```python
[expresión for item in iterable if condición]
```

````python
even_numbers = [i for i in range(21) if i % 2 == 0]
print(even_numbers)
````

Piensalo de la siguiente forma:

```python
[ (for) if condition ] --> code1 if condition
```
Si la condicion se cumple devuelve el valor iterado correspondiete.


Se genera `range` y el `item` toma el valor inicial, este valor inicial entra a la siguiente estructura:


```python
item if condition
```
si se cumple la condición, se devuelve el item. En otras palabras es una lista filtrada.


## Condicional ternario



```python
[valor_si_verdadero if condición else valor_si_falso for item in iterable]
```



````python
etiquetas = ['par' if x % 2 == 0 else 'impar' for x in range(5)]
etiquetas
````

Piensalo de esta forma:

```python
[(if else) for item in iterable]  --> [expresión for item in iterable]
```
Se genera el `iterable`, `item` toma el valor inicial. `item` entra a la siguiente estructura:


```python
code1 if condicion(item) else code2
```

Como puedes oberservar no esta filtrando, esta transformando el valor de `item`.

A la lista ingresa `code1` o `code2` dependiendo de el cumplimiento de la condicion.





````python
# Flattening a three dimensional array
list_of_lists = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

flattened_list = [ number for row in list_of_lists for number in row]
print(flattened_list)    # [1, 2, 3, 4, 5, 6, 7, 8, 9]
````

#Funcion `lambda:`

Una función lambda es una manera rápida y compacta de definir funciones pequeñas y anónimas (es decir, sin nombre).



```python
lambda argumentos: expresión

```



````python
suma = lambda a,b: a+b
print(suma(1,2))
````

````python
#4
palabras = ['sol', 'montaña', 'río', 'universo', 'luz']

string_larga = filter(lambda x: x if len(x)>5 else '', palabras)

list(string_larga)

#solucion optimizada:
string_larga = filter(lambda x: len(x) > 5, palabras)
````
