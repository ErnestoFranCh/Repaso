## `timestamp()`

La función `.timestamp()` convierte un objeto `datetime` en un número de tipo `float` que representa la cantidad de segundos desde el 1 de enero de 1970 a las 00:00:00 UTC (conocido como Epoch time o Unix timestamp).

`timestamp()` ---> `datetime` : `datetime`.`fromtimestamp`(`stamp`)

Sirve para:

- Guardar fechas de forma compacta (bases de datos, archivos, APIs).

- Comparar tiempos.

- Calcular duración entre eventos fácilmente.

- Sincronización con sistemas que usan Unix Time.

````python
from datetime import datetime
stamp = now.timestamp()

print(stamp)
````

````python
date = datetime.fromtimestamp(stamp)
print(date)
````

## `time`

````python
from datetime import time
# time(hour = 0, minute = 0, second = 0)
a = time()
print("a =", a)

# time(hour, minute and second)
b = time(10, 30, 50)
print("b =", b)

# time(hour, minute and second)
c = time(hour=10, minute=30, second=50)
print("c =", c)

# time(hour, minute, second, microsecond)
d = time(10, 30, 50, 200555)
print("d =", d)
````

## Diferencia de tiempo

````python
from datetime import datetime, date

today = date(year=2019, month=12, day=5)

new_year = date(year=2020, month=1, day=1)

time_left_for_newyear = new_year - today
# Time left for new year:  27 days, 0:00:00
print('Time left for new year: ', time_left_for_newyear)

t1 = datetime(year = 2019, month = 12, day = 5,
              hour = 0, minute = 59, second = 0)

t2 = datetime(year = 2020, month = 1, day = 1,
              hour = 0, minute = 0, second = 0)

diff = t2 - t1
print('Time left for new year:', diff) # Time left for new year: 26 days, 23: 01: 00
````

### Diferencia de tiempo con `timedelta`

`timedelta` representa una duración o una diferencia de tiempo, es decir, el resultado de restar dos objetos datetime o una cantidad fija de tiempo.

````python
from datetime import timedelta

delta = timedelta(days=3, hours=5, minutes=30)
print(delta)
````

````python
from datetime import datetime

hoy = datetime.now()
en_5_dias = hoy + timedelta(days=5)
ayer = hoy - timedelta(days=1)

print("Hoy:      ", hoy)
print("En 5 días:", en_5_dias)
print("Ayer:     ", ayer)
````

````python
t1 = timedelta(weeks=12, days=10, hours=4, seconds=20)
t2 = timedelta(days=7, hours=5, minutes=3, seconds=30)
t3 = t1 - t2
print("t3 =", t3)
````

## `strftime`
`strftime` significa "string format time".
Es un método que convierte un objeto datetime o date en un string, usando un formato personalizado que tú defines.

````python
from datetime import datetime
#sin strftime
new_year = datetime.now()
print(new_year)

print('############################\n')

day = new_year.day
month = new_year.month
year = new_year.year
hour = new_year.hour
minute = new_year.minute
second = new_year.second

print(day, month, year, hour, minute)
print(f'{day}/{month}/{year}, {hour}:{minute}')
````

````python
from datetime import datetime

now = datetime.now()

t = now.strftime("%H:%M:%S")
print("time:", t)

time_one = now.strftime("%m/%d/%Y, %H:%M:%S")
# mm/dd/YY H:M:S format
print("time one:", time_one)


time_two = now.strftime("%d/%m/%Y, %H:%M:%S")
# dd/mm/YY H:M:S format
print("time two:", time_two)

time_three = now.strftime("%a %d %b")
print("time three:", time_three)

time_four = now.strftime("%A %d. %B %Y")
print("time four:", time_four)
````

## `string` --> `datetime`

Para convertir una fecha tipo `str` a tipo `datetime`, debe ser una cadena que siga un formato reconocido, y normalmente se usa la función:



```python
from datetime import datetime
datetime.strptime()
```

#Estructura de la cadena.

| Ejemplo de Cadena       | Formato              |
|-------------------------|----------------------|
| "2025-04-16"            | "%Y-%m-%d"           |
| "16/04/2025"            | "%d/%m/%Y"           |
| "16-04-2025 14:30"      | "%d-%m-%Y %H:%M"     |
| "April 16, 2025"        | "%B %d, %Y"          |
| "16 Apr 2025"           | "%d %b %Y"           |



````python
#ej
from datetime import datetime

fecha_str = "16/04/2025"
fecha_dt = datetime.strptime(fecha_str, "%d/%m/%Y")
print(fecha_dt)

fecha2_str = "15-03-1998"
fecha2_dt = datetime.strptime(fecha2_str,"%d-%m-%Y").date()
print(fecha2_dt)
````
